**Write out the mojo kernel**

```python
%%writefile block_kernel.mojo

from gpu import block_idx, thread_idx
from gpu.host import DeviceContext

fn block_kernel():
    print("Block: [", 
            block_idx.x, block_idx.y, block_idx.z, "]",
          "Thread: [",
             thread_idx.x, thread_idx.y, thread_idx.z, "]")

fn main() raises:
    var ctx = DeviceContext()
    ctx.enqueue_function[block_kernel](grid_dim=(2,2), block_dim=(2, 2))
    ctx.synchronize()
    print("After kernel execution")
```
